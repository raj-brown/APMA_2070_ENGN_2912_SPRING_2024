# Gradient: PyTorch vs TensorFlow

```python
# In TensorFlow
import tensorflow as tf
import numpy as np

x = 3.0
x_tensor = tf.constant(3.0)
with tf.GradientTape(persistent=True) as f:
    f.watch(x_tensor)
    with tf.GradientTape(persistent=True) as g:
        g.watch(x_tensor)
        y = tf.exp(x_tensor)
    dy_dx = g.gradient(y, x_tensor)
d2y_dx2 = f.gradient(dy_dx, x_tensor)


print(f"In TF2 First derivative of exp(x): {d2y_dx2} and actual gradient: {np.exp(x)}")
```

```python
# In TensorFlow
import torch
import numpy as np
x_fl = 3.0
x = torch.tensor(3.0, requires_grad = True)
y = torch.exp(x)
y.backward()  
dy_dx = x.grad.data #=: dy/dx 
print(f"First derivative of exp(x): {dy_dx} and actual gradient: {np.exp(x_fl)}")
```
